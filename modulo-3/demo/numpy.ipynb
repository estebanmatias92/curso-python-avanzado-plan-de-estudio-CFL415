```python
#import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
indices = np.indices((3, 3, 3))

fig = plt.figure()

# Plotting the first array
ax = fig.add_subplot(131, projection='3d')
ax.scatter(indices[0], indices[1], indices[2], c=indices[0].flatten())
ax.set_title('X indices')

# Plotting the second array
ax = fig.add_subplot(132, projection='3d')
ax.scatter(indices[0], indices[1], indices[2], c=indices[1].flatten())
ax.set_title('Y indices')

# Plotting the third array
ax = fig.add_subplot(133, projection='3d')
ax.scatter(indices[0], indices[1], indices[2], c=indices[2].flatten())

plt.show()
```

```python
# Create a 4x4x4 array for example
array_3d = np.arange(1, 65).reshape(4, 4, 4)

# Get the indices for the space diagonals
indices = np.arange(array_3d.shape[0]).view()
space_diag0 = (indices, indices, indices)
space_diag1 = (indices, indices, indices[::-1])
space_diag2 = (indices[::-1], indices, indices)
space_diag3 = (indices[::-1], indices, indices[::-1])

# Print the space diagonals
print("Space diagonal 1:", array_3d[space_diag0])
print("Space diagonal 2:", array_3d[space_diag1])
print("Space diagonal 3:", array_3d[indices[::-1], indices, indices])
print("Space diagonal 4:", array_3d[indices[::-1], indices, indices[::-1]])

```

```python
# Create a figure for plotting
fig = plt.figure()

# Add 3D subplot
ax = fig.add_subplot(111, projection='3d')

# Plot the points in the array
x, y, z = np.where(array_3d >= 1)  # Get the indices of the elements

# Use z (the layer index) as the color index
colors = plt.cm.viridis(z / max(z))

ax.scatter(x, y, z, c=colors)

# Plot the space diagonals
ax.plot(*space_diag0, color='red')  # Diagonal from (0,0,0) to (3,3,3)
ax.plot(*space_diag1, color='blue')  # Diagonal from (0,3,0) to (3,0,3)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()


```
